```python
import pandas as pd

read_file = pd.read_csv ('Hamming.txt')

print(read_file)
df2 = read_file.groupby(['Errors'])['Errors'].count()

df2.plot(x='Errors',
        kind='bar',
        stacked=False,
        title='Hamming. Error de 1 en 100')
```
